```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les résultats CSV
df = pd.read_csv('yolov5/runs/train/epi_yolov5/results.csv')

# Tracer les métriques
metrics = ['train/box_loss', 'train/obj_loss', 'train/cls_loss', 
           'metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP_0.5', 'metrics/mAP_0.5:0.95']

plt.figure(figsize=(12, 6))
for metric in metrics:
    if metric in df.columns:
        plt.plot(df[metric], label=metric)

plt.xlabel("Epoch")
plt.ylabel("Valeur")
plt.title("Courbes d'apprentissage YOLOv5")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
